<font size='5'> **Kmeans Clustering**  </font>

<font size='4'> **Definition**  </font>

K-means clustering is a simple unsupervised learning algorithm that is used to solve clustering problems. It follows a simple procedure of classifying a given data set into a number of clusters, defined by the letter "k," which is fixed beforehand. The clusters are then positioned as points and all observations or data points are associated with the nearest cluster, computed, adjusted and then the process starts over using the new adjustments until a desired result is reached.

K-means clustering has uses in search engines, market segmentation, statistics and even astronomy.

```python
df.head()
```

```python
clustering_data=df.iloc[:,3:]
clustering_data.head()
```

<font size='4'>**Visualization between annual income and spending score**</font>

```python
sns.scatterplot(x='annual_income',y='spending_score',hue='gender',data=df)
```

**Obsevation**

1.It can be observed that by seeing annual income and spending score we can roughly divide the data in 5 groups.

<font size='5'>**Determining the number of clusters**</font>

“Elbow” method  helps to select the optimal number of clusters by fitting the model with a range of values for K. If the line chart resembles an arm, then the “elbow” (the point of inflection on the curve) is a good indication that the underlying model fits best at that point.

```python
from sklearn.cluster import KMeans
wcss=[]    #wcss stands for Within-Cluster Sum of Square
for i in range(1,21):
  kmeans = KMeans(n_clusters=i,init='k-means++',max_iter=300,n_init=10)   
  #init = 'kmeans++' beacuse it  selects initial cluster centers for k-mean clustering in a smart way to speed up convergence
  #max_iter   this is the maximum number of iteration that can be done to find final cluster
  #n init = which is number of times k means algorithm will run with different initial centroids
  kmeans.fit(clustering_data)
  wcss.append(kmeans.inertia_) 
```

```python
fig, av = plt.subplots(figsize=(15,7))
av = plt.plot(range(1,21),wcss, linewidth=2, color="red", marker ="8")
plt.axvline(x=5, ls='--')
plt.ylabel('WCSS',fontsize = 20)
plt.xlabel('Number of Clusters',fontsize = 20)
plt.title('The Elbow Method',fontsize = 30)
```

**Obervation**

By using elbow method we csn say that optimal number of cluster is k=5

```python
clustering_data = clustering_data.values

#applying kmeans to mall dataset

kmeans = KMeans(n_clusters=5,init='k-means++',max_iter=300,n_init=10)   
ykmeans = kmeans.fit_predict(clustering_data)
```

```python
ykmeans
```

```python
plt.figure(figsize=(15,8))

plt.scatter(clustering_data[ykmeans==0,0] , clustering_data[ykmeans==0,1], s=70,edgecolor='black', linewidth=0.3, c='red', label='Target')
plt.scatter(clustering_data[ykmeans==1,0] , clustering_data[ykmeans==1,1], s=70,edgecolor='black', linewidth=0.3, c='blue', label='Careless')
plt.scatter(clustering_data[ykmeans==2,0] , clustering_data[ykmeans==2,1], s=70,edgecolor='black', linewidth=0.3, c='green', label='Sensible')
plt.scatter(clustering_data[ykmeans==3,0] , clustering_data[ykmeans==3,1], s=70,edgecolor='black', linewidth=0.3, c='cyan', label='Standard')
plt.scatter(clustering_data[ykmeans==4,0] , clustering_data[ykmeans==4,1], s=70,edgecolor='black', linewidth=0.3, c='magenta', label='Careful')

plt.scatter(x=kmeans.cluster_centers_[:, 0], y=kmeans.cluster_centers_[:, 1], s = 120, c = 'yellow', label = 'Centroids',edgecolor='black', linewidth=0.3)

plt.xlabel('Annual Income',fontsize=20)
plt.ylabel('Spending score',fontsize=20)
plt.title('Cluster of customers',fontsize=30)
plt.legend()
plt.grid()
```

<font size='5'>**Results** </font>

From the above graph we get a clear idea about the distribution of data on basis of annual income and spending score.Kmeans have divided the data in 5 clusters they are as follows: 

**1.Careful**

**2.Standard**

**3.Sensible**

**4.Careless**

**5.Target** 

Now we will analyze each type of customer so that you as a **Mall owner** can decide on which type of **customers** to focused on
